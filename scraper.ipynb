```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
from pprint import pprint
from html_table_parser.parser import HTMLTableParser
import pandas as pd


def url_get_contents(url):
    req = urllib.request.Request(url=url)
    f = urllib.request.urlopen(req)
    return f.read()
 
xhtml = url_get_contents('https://www.treasuredata.com/list-of-cookies/').decode('utf-8')
 
p = HTMLTableParser()
p.feed(xhtml)
pprint(p.tables)
print("\n\nPANDAS DATAFRAME\n")
print(pd.DataFrame(p.tables))

url = 'https://www.worldometers.info/coronavirus/'
page = requests.get(url)

soup = BeautifulSoup(page.text, 'lxml')
table1 = soup.find('table', id = 'ot-scrn-rdr')
headers = []
for i in table1.find_all('th'):
    title = i.text
    headers.append(title)

mydata = pd.DataFrame(columns = headers)

for j in table1.find_all('tr')[1:]:
    row_data = j.find_all('td')
    row = [i.text for i in row_data]
    length = len(mydata)
    mydata.loc[length] = row

# Drop and clearing unnecessary rows
mydata.drop(mydata.index[0:7], inplace = True)
mydata.drop(mydata.index[222:229], inplace = True)
mydata.reset_index(inplace = True, drop = True)

mydata.drop('#', inplace = True, axis = 1)
```

```python
import requests
import lxml
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.komatsuforest.com/list-of-cookies'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')

table = soup.find('table', {'class':'table table-striped'})
headers = []

for i in table.find_all('th'):
    headers.append(i.text.strip())

df = pd.DataFrame(columns = headers)
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data

url = 'https://www.iwm.org.uk/corporate/policies/cookies/cookie-list-main'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'lxml')

table = soup.find('div', {'class':'text-only__content'})
headers = []
headers = ['Cookie', 'Duration', 'Description']
df3 = pd.DataFrame(columns = headers)

database = []
for row in table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    database.append(row_data)
    if len(row_data) == 4:
        row_data.pop(2)
        df3.loc[length] = row_data
    elif len(row_data) == 3:
        length = len(df3)
        df3.loc[length] = row_data
database = [database.pop(index) for index, value in enumerate(database)]
for row in database:
    row.append('tracking')

LOAD_DATABASE = []
with open(r'database.txt', 'w') as fp:
    for row in database:
        if len(row) == 4:
            update_row = ''.join('INSERT INTO FILTER_LIST (cookie, domain, description, type) VALUES(') + row[0], row[1], row[2], row[3] 
            LOAD_DATABASE.append(update_row)
            txt = ''.join('INSERT INTO FILTER_LIST (cookie, domain, description, type) VALUES(') + '\'' + row[0] + '\'' + ', ' + '\'' + row[1] + '\'' + ', ' + '\'' + row[2] + '\'' + ', ' + '\'' + row[3] + '\'' + ')' + ';'  
            print(txt)
            fp.write('%s\n' %txt)
LOAD_DATABASE
```
